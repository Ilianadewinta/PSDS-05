```python
def fungsi (a,b) :
  keliling = 2*(a+b)
  luas = a*b
  print('Panjang persegi panjang adalah :', a, 'cm')
  print('Lebar persegi panjang adalah :', b, 'cm')
  print('Keliling persegi panjang adalah :', keliling, 'cm')
  print('Luas persegi panjang adalah :', luas, 'cm^2')

fungsi(10,8)
```
